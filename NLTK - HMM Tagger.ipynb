### Libraries 

```python
from nltk.tag import HiddenMarkovModelTagger
from nltk.tag import HiddenMarkovModelTrainer
from pprint import pprint
from nltk.probability import LidstoneProbDist
```

### Load Corpus 

```python
fname = "Corpus.txt"
fopen = open(fname, 'r')
fdata = fopen.readlines()

pprint(fdata[:5])
```

```python
sentences = []
temp = []
for word in fdata:
    data = word.replace('\n','').split('\t')
    if(len(data) == 1):
        sentences.append(temp)
        temp = []
    else:
        temp.append((data[0],data[1]))
```

### Training 

```python
trainer = HiddenMarkovModelTrainer()
tagger = trainer.train_supervised(sentences, estimator=lambda fd, bins: LidstoneProbDist(fd, 0.1, bins))
```

### Test 

```python
tagger.test(sentences, verbose=True)
```
